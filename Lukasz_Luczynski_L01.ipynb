```python
import pandas as pd
import csv
```

```python
# Zczytywanie i uporządkowywanie danych
data = []
n = 1
with open('earthquake_data.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        data.append(row)
data_dict = {}
for i in range(len(data[0])):
    data_dict[data[0][i]] = [row[i] for row in data[1:]]
```

```python
# Stworzenie DataFrame'u
df = pd.DataFrame(data_dict)
df
```
