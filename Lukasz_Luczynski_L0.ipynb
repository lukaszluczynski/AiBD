Zdefiniuj poniższą funkcję i sporządź jej wykres dla argumentów z danego przedziału:

$f(x)=x^2+5$

    $x>-1 oraz x<1$
    $x>-6 oraz x<6$
    $x>0 oraz x<5$


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas
```

```python
def f(x):
    return x**2+5
```

```python
Blist=[[-1, 1], [-6, 6], [0, 5]]
for B in Blist:
    X = np.linspace(B[0], B[1])
    FX = f(X)
    plt.plot(X, FX)
    plt.grid()
    plt.xlabel("oś X")
    plt.legend(["Wykres"])
    plt.ylabel("oś Y")
    plt.title("Tytuł wykresy")
    plt.show()
```
